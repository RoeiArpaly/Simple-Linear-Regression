# The Simple Mathematics Behind The Simple Linear Regression

Imports:

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='darkgrid', rc={'axes.linewidth':1, 'axes.edgecolor':'black'})
%matplotlib inline
```

Creating some simple data to work with:

```
X = np.array([range(0,91,10)]).reshape(10,1)
y = np.array([400,450,555,500,400,550,600,700,650,800]).reshape(10,1)
```

Plotting the data:

```
sns.scatterplot(x=X[:,0], y=y[:,0], cmap='Blues', edgecolor='black', linewidth=1, s=50)
plt.xlim(-5,100), plt.ylim(-50,1000)
plt.show()
```

## Manual construction of the Simple Linear Regression

Calculating the differences between the independent variable (x) and its mean and the dependent variable (y) and its mean. 

```
X_diff = X - X.mean()
y_diff = y - y.mean()
```

Multiplying the differences of x and y.

```
Xy_diff = X_diff * y_diff
```

Calculating the squared difference of the independent variable (x) to overcome to affect of the negative values.

```
X_diff_squared = X_diff**2
```

Then, getting the mean out of the equations.

```
Sxx = X_diff_squared.mean()
Sxy = Xy_diff.mean()
```

Finally, getting the coefficient and the constant of the equation.

```
Coefficient = Sxy/Sxx
Constant = y.mean() - Coefficient * X.mean()
```

Construct the Simple Linear Regression equation:

```
print(f'Y = {Constant} + {Coefficient}*(X)')
```

Plotting the Regression line:

```
sns.scatterplot(x=X[:,0], y=y[:,0], edgecolor='black')
sns.lineplot(x=X[:,0], y=Constant+Coefficient*X[:,0], color='blue')

plt.title('Simple Linear Regression', fontsize=16)
plt.xlabel('Independent variable'), plt.ylabel('Dependent variable')
plt.xlim(-5,100), plt.ylim(-50,1000)
plt.show()
```

## Building the Simple Linear Regression using Scikit-learn

2 lines of code!

```
from sklearn.linear_model import LinearRegression
regressor = LinearRegression().fit(X, y)
```

```
sns.scatterplot(x=X[:,0], y=y[:,0], edgecolor='black')
sns.lineplot(x=X[:,0], y=regressor.predict(X)[:,0], color='blue')

plt.title('Simple Linear Regression', fontsize=16)
plt.xlabel('Independent variable'), plt.ylabel('Dependent variable')
plt.xlim(-5,100), plt.ylim(-50,1000)
plt.show()
```
